##Imputing missing brands:

```python
brands = brand_df['Brand'].values
brands_len=[]
for brand in brands:
    brands_len.append(len(brand.split()))

#max no. of words in  any brand
print(max(brands_len))
#Average no. of words in any brand
print(np.array(brands_len).mean())
```

```python
#Function to fill in missing values of brand

def fill_brand(df,absent_products,brands):
    for prod in absent_products:
        for title in df[df['product_uid'] == prod]['product_title'].unique():
            if title in brands:
                df.loc[df.product_uid == prod, 'Brand'] = title
            else:
                word = title.split()
                if word[0] not in brands:
                    string = word[0]
                    if len(word) >6:
                        for combined in range(1, 6):
                            string = string + ' ' +word[combined]
                            if string in brands:
                                df.loc[df.product_uid == prod, 'Brand'] = string
                                break
                            else:
                                continue
                    else:
                        for combined in range(1, len(word)):
                            string = string + ' ' +word[combined]
                            if string in brands:
                                df.loc[df.product_uid == prod, 'Brand'] = string
                                break
                            else:
                                continue
                else:
                    df.loc[df.product_uid == prod, 'Brand'] = word[0]

```

```python
#filling brand missing values for train
fill_brand(train_all, train_brand_absent, brands)
```

```python
#filling brand missing values for test
fill_brand(test_all, test_brand_absent, brands)
```

##Basic Preprocessing:

```python
# preprocessing funtion 
def clean_text(sent):
    x = sent

    #punctuations and special chars
    x = re.sub('[^a-zA-Z0-9_ ]',' ', x)
    
    #capital letters seperation
    x = ' '.join(re.findall(r'[A-Z]?[^A-Z\s]+|[A-Z]+', x))

    #lowercase
    x = str(x).lower()
    #html tags
    x = re.sub('<.*?>',' ',x)

    
    #normalize units before removing '.' (in. --> inch)

    string = re.findall('\d+ *in\.{0,1} *',x)
    if len(string)!=0:
        for st in string:
            x = x.replace(st, ' inch ') #since 'in' is a preposition we check if it is follwed by a number
    x = re.sub('(inches)', ' inch ', x)
    string = re.findall('\d+ *cu\.{0,1} *',x)#cubic
    if len(string)!=0:
        for st in string:
            x = x.replace(st, ' cubic ')
    string = re.findall('\d+ *m\.{0,1} *',x)#metre
    if len(string)!=0:
        for st in string:
            x = x.replace(st, ' metre ')
    string = re.findall('\d+ *cm\.{0,1} *',x)#centimetre
    if len(string)!=0:
        for st in string:
            x = x.replace(st, ' centimetre ')
    x = re.sub('( centimetres | centi-metres )',' centimetre ',x) 
    x = re.sub('( metres )',' metre ',x)
    x = re.sub('( ft | fts | feets | foot | foots )','feet',x)
    x = re.sub('( gal | gals | galon )',' gallon ',x)
    x = re.sub('( yds | yd | yards )',' yard ',x)
    x = re.sub('( oz | ozs | ounces | ounc ) ',' ounce ',x)
    x = re.sub('( lb | lbs | pounds )',' pound ',x)
    x = re.sub('( squares | sq )',' square ',x)

    

    #Seperating numbers from letters (8x --> 8 x)
    x = ' '.join([str(ele) for ele in re.split(r'(\d+)', x)])

    #removing extra spaces
    x = re.sub(' +', ' ', x)

    return x
```

##Stopword removal and stemming

```python
import nltk
nltk.download('stopwords')
```

```python
#Snowball stemmer
snow = SnowballStemmer(language='english')

#function to remove stopwords and perform stemming
def preproecessing(sent): 
    sent = sent.replace('_', ' _ ')
    words = sent.split()
    #not including stopwords 
    words = [w for w in words if not w in set(stopwords.words('english'))]
    #coverting to stem form
    words = [snow.stem(word) for word in words]
    return ' '.join(words)

```

## Common words in search and document : 


5. common words in search and title
6. common words in search and desc


```python
#common words
def common_words(search, doc):
    words = []
    for i in range(len(search)):
        search_words = set(search[i].split())
        doc_words = set(doc[i].split())
        words.extend(list(search_words & doc_words))
    return words
```

```python
common_S_T = common_words(list(train_all.search_term_corrected.values), list(train_all.product_title.values))
common_S_D = common_words(list(train_all.search_term_corrected.values), list(train_all.product_description.values))
common_S_B = common_words(list(train_all.search_term_corrected.values), list(train_all.Brand.apply(str).values))
```

```python
#len of common words
def len_common(search, doc):
    length =[]
    for i in range(len(search)):
        search_words = set(search[i].split())
        doc_words = set(doc[i].split())
        length.append(len(search_words & doc_words))
    return length
```

##Query last word :
Usually last term has most importance amongst all in a search term:
7. last_in_title
8. last_in_desc
9. last_in_brand

```python
def last_in_doc(search, doc):
    last = []
    for i in range(len(search)):
        last_term = search[i].split()[-1]
        if last_term in doc[i].split():
            last.append(1)
        else:
            last.append(0)
    return last
```

```python
#last term search and product title
train_02['last_in_title'] = last_in_doc(list(train_all.search_term_corrected.values), list(train_all.product_title.values))
#last term search and product desc
train_02['last_in_desc'] = last_in_doc(list(train_all.search_term_corrected.values), list(train_all.product_description.values))

#last term search and product brand
train_02['last_in_brand'] = last_in_doc(list(train_all.search_term_corrected.values), list(train_all.Brand.apply(str).values))

train_02.head()
```
